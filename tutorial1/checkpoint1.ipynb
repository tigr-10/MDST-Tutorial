<h2> Data Cleanup </h2>

Import `numpy`, `pandas`, and `matplotlib`.

(Feel free to import other libraries!)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Load in the "states_edu.csv" dataset and take a look at the head of the data

```python
#load in data
df = pd.read_csv("../data/states_edu.csv")
df.head()
```

You should always familiarize yourself with what each column in the dataframe represents. Read about the states_edu dataset here: https://www.kaggle.com/noriuk/us-education-datasets-unification-project

Use this space to rename columns, deal with missing data, etc. _(optional)_

```python
df["ENROLL"] = df["ENROLL"].fillna(df["GRADES_ALL_G"])
```

```python
#let's see the distribution of AVG_MATH_8_SCORE to see whether or not we should fill na's with mean or median
df.AVG_MATH_8_SCORE.plot.hist(title="Distribution of 8th Grade Math Scores", edgecolor="black")

#since the distribution is left skewed, let's fill in na's with median
#but, we should separate by state

# Calculate the median AVG_MATH_8_SCORE for each state
state_medians = df.groupby('STATE')['AVG_MATH_8_SCORE'].median()

# Define a function to fill missing values with the state-specific median
def fill_missing_with_state_median(row):
    if pd.isna(row['AVG_MATH_8_SCORE']):  # Check if the score is NaN
        return state_medians[row['STATE']]  # Return the state-specific median
    else:
        return row['AVG_MATH_8_SCORE']  # Return the original value if it's not NaN

# Apply the function to the DataFrame
df['AVG_MATH_8_SCORE'] = df.apply(fill_missing_with_state_median, axis=1)
```

```python
df["AVG_MATH_8_SCORE"].isna().sum()
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['AVG_MATH_4_SCORE','AVG_READING_4_SCORE','AVG_READING_8_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.
